```python
# Soal 4 dan 5
#import lah library Numpy
''' Dengan memanfaatkan library numpy Buatlah matriks dengan menggunakan 
Library Numpy dengan panjang elemen adalah 32 dan matriks memiliki orde 4x8'''

import numpy as np

# Membuat array 1D dengan panjang elemen 32
array_1d = np.arange(32)

# Mengubah array 1D menjadi matriks 4x8
matriks_4x8 = np.reshape(array_1d, (4, 8))

# Menampilkan matriks
print(matriks_4x8)

```
