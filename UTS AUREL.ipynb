```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# membuat dataframe dari data mahasiswa
fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]
info_mahasiswa = pd.DataFrame({"Fakultas": fakultas, "Jumlah Mahasiswa": jumlah_mahasiswa, "Akreditasi": akreditasi})

# membuat grafik
sns.barplot(x="Fakultas", y="Jumlah Mahasiswa", data=info_mahasiswa, hue="Akreditasi")

# menambahkan judul pada grafik
plt.title('Jumlah Mahasiswa per Fakultas')

# menampilkan grafik
plt.show()
```

```python
import pandas as pd

# membuat dictionary untuk biaya mahasiswa
biaya = {'Ani': 30000, 'Budi': 35000, 'Joni': 20000, 'Jono': 25000, 'Lono': 15000}

# membuat dataframe dari data mahasiswa
data = {'Mahasiswa': ['Ani', 'Budi', 'Joni', 'Jono', 'Lono'],
        'Senin': [2, 0, 0, 0, 0],
        'Selasa': [0, 3, 0, 0, 0],
        'Rabu': [0, 0, 4, 0, 0],
        'Kamis': [0, 0, 0, 0, 1],
        'Jumat': [0, 0, 0, 2, 0],
        'Sabtu': [5, 0, 0, 0, 0],
        'Minggu': [0, 2, 0, 0, 0]}
df = pd.DataFrame(data)

# menghitung total biaya per hari
for hari in df.columns[1:]:
    df[hari] = df[hari] * biaya[df['Mahasiswa'][df[hari].to_numpy().nonzero()[0][0]]]

# a) rata-rata mahasiswa datang pada minggu ini
rata_rata = df.iloc[:, 1:].astype(bool).sum(axis=1).mean()
print("Rata-rata mahasiswa datang pada minggu ini:", rata_rata)

# b) kapan biaya tertinggi terjadi
tertinggi = df.iloc[:, 1:].sum().idxmax()
print("Biaya tertinggi terjadi pada hari", tertinggi)

# c) hari apa biaya lebih dari 110000
lebih_dari_110000 = df.iloc[:, 1:].columns[df.iloc[:, 1:].sum() > 110000].tolist()
print("Biaya lebih dari 110000 terjadi pada hari", lebih_dari_110000)

# d) siapa yang paling banyak datang ke kampus
paling_banyak = df.iloc[:, 1:].astype(bool).sum(axis=1).idxmax()
print("Mahasiswa yang paling banyak datang ke kampus adalah", df['Mahasiswa'][paling_banyak])

# e) siapa yang datang pada hari minggu
minggu = df.iloc[:, 7][df.iloc[:, 7] != 0].index[0]
print("Mahasiswa yang datang pada hari Minggu adalah", df['Mahasiswa'][minggu])

# f) biaya tertinggi dan terendah
tertinggi = df.iloc[:, 1:].max().max()
terendah = df.iloc[:, 1:].min().min()
print("Biaya tertinggi adalah", tertinggi, "dan biaya terendah adalah", terendah)

# g) frekuensi datang tertinggi dan terendah
frekuensi_tertinggi = df.iloc[:, 1:].astype(bool).sum().max()
frekuensi_terendah = df.iloc[:, 1:].astype(bool).sum().min()
print("Frekuensi datang tertinggi adalah", frekuensi_tertinggi, "dan frekuensi datang terendah adalah", frekuensi_terendah)
```
